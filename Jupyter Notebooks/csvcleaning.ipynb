```python
# Dependencies
import pandas as pd
```

```python
#file to load into notebook
file= "../Data/voting-age.csv"

#store the purchase_data file into a data frame
age_states = pd.read_csv(file)

#store the ages in a dataframe
age_df=pd.DataFrame(age_states)
age_df.head(5)
```

```python
 #grab the first row for the header.
age_header = age_df. iloc[0]

#take the data less the header row.
age_tables = age_df[1:] 

#set the header row as the df header.
age_tables . columns =age_header 

#select only the columns that we want from the csv
i=age_tables.iloc[:,0:14]

#select all rows except for the last two which are NANs
ages_df=i[1:313]
ages_df=ages_df.rename(columns = {'STATE': 'state','Age':'age','Total Citizen Population':'total_citizen_population',
                          'Percent registered\n(Citizen)':'percent_registered_citizen',
                          'Percent voted\n(Citizen)':'percent_voted_citizen'})
```

```python
#view rows with "B"
ages_df[ages_df.values == 'B']
#replace the B with Nan and save as a new dataframe
clean_data_age=ages_df.replace(to_replace ="B", value ="Nan")
clean_data_age
```

```python
clean_data_age_df=clean_data_age.set_index("state")
clean_ages=clean_data_age_df[['age','total_citizen_population','percent_registered_citizen','percent_voted_citizen']]
clean_ages
```

```python
clean_ages.to_csv('../Clean_Data/voting_ages.csv')
```

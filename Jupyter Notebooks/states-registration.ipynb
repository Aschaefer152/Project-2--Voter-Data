```python
# Dependencies
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
#create the response with the url that we visit
url="https://www.ncsl.org/research/elections-and-campaigns/same-day-registration.aspx#_Toc522006760"
response=requests.get(url)

#create the beautiful soup object
soup = bs(response.text,'html')
#print(soup.prettify())

#print the tables on the webpage
tables=pd.read_html(url)
voting_tables=tables[0]
voting_tables.head(5)
```

```python
#grab the first row for the header.
voting_tables_header = voting_tables. iloc[0] 

 #take the data less the header row.
voting_tables = voting_tables[1:]

#set the header row as the df header.
voting_tables. columns = voting_tables_header 
voting_tables.head()
```

```python
voting_tables
```

```python
# voting_tables.to_csv("Data/states-registration.csv")
state_names=voting_tables['State']
```

```python
#create an empty list
states=[]
```

```python
#add all the states that have early or same day voter registration legislation
state = input("Enter the state name")
for i in state_names:
    if state in i:
        states.append(state)
        
print(states) 
```

```python
#create a column of state names with the states list
voting_tables['state']=states
voting_tables
```

```python
#change the two early voting state to match their legislation
voting_tables.loc[(voting_tables['State Names'] == 'North Carolina'),'Registration']='Early Voting'
voting_tables.loc[(voting_tables['State Names'] == 'New Mexico'),'Registration']='Early Voting'
```

```python
#re-order the table and only select the columns that we want
voting_tables_df=voting_tables[['state','Registration']]
vote_df=voting_tables_df.rename(columns = {'Registration':'voting_registration'})
# vote_df.set_index('state').to_csv('Clean_Data/registration_table.csv')
vote_df
```

```python
#get the lat long csv
file= "../Data/States - Lat Long Data.csv"

#store the purchase_data file into a data frame
lat_long_df = pd.read_csv(file)
lat_long_df.head(5)
```

```python
#rename the columns
lat_long_df=lat_long_df.rename(columns = {'City':'state','State':'abbr',
                                          'Latitude':'latitude',
                                          'Longitude':'longitude'})
reg_merge=lat_long_df.merge(voting_tables,how="left",on='state')
reg_final=reg_merge.fillna('none')
lower_r_df=reg_final.rename(columns = {'Registration':'registration'})
lower_r_df[['state','registration']].set_index('state').to_csv('../Clean_Data/registration_table.csv')
```
